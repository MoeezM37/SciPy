```python
import numpy as np

nums = np.random.randint(-10, 10, size=(6,6))
print("Original array is \n\n", nums)

# Cutting out 1st line
nums1 = nums.copy()
nums1 = np.delete(nums1, 0, 0)
print("\nCutting out 1st line: \n\n", nums1)

# Multiplying 5th row by 2
nums2 = nums.copy()
nums2[4,:] *= 2
print("\nMultiplying 5th row by 2: \n\n", nums2)

# Cutting odd columns
nums3 = nums.copy()
nums3 = np.delete(nums3, [0,2,4], 1)
print("\nCutting odd columns: \n\n", nums3)

# Cutting out 3x3 matrix
nums4 = nums.copy()
nums4 = nums4[0:3,0:3]
print("\nCutting out 3x3 matrix: \n\n", nums4)

# Setting all negative numbers to 0
nums5 = nums.copy()
for i in range(nums5.shape[0]):
    for j in range(nums5.shape[1]):
        if nums5[i,j] < 0:
            nums5[i,j] = 0
print("\nSetting all negative numbers to 0: \n\n", nums5)

# Cutting 5th column
nums6 = nums.copy()
nums6 = np.delete(nums6, 4, 1)
print("\nCutting 5th column: \n\n", nums6)

# Cutting even rows
nums7 = nums.copy()
nums7 = np.delete(nums7, [1,3,5], 0)
print("\nCutting even rows: \n\n", nums7)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

pic = plt.imread("sky.png")
print("\nOriginal picture: ")
plt.imshow(pic)
plt.show()

pic_b = pic[:,:,2]
pic_gray = np.mean(pic, axis=2)
pic_check1 = pic_b/pic_gray
print("\nCheck picture 1 (relative blue component): ")
plt.imshow(pic_check1)
plt.colorbar()
plt.show()

pic_check2 = pic_check1.copy()
for i in range(pic_check2.shape[0]):
    for j in range(pic_check2.shape[1]):
        if pic_check2[i,j] > 1.3:
            pic_check2[i,j] = 1
        else:
            pic_check2[i,j] = 0
print("\nCheck picture 2 (cloudiness): ")
plt.imshow(pic_check2)
plt.colorbar()
plt.show()

cloudiness = np.mean(pic_check2)
print("\nThe degree of cloudiness in the provided image is: ", cloudiness*100, "%")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math

def A(eta, A_s, D):
    return ((A_s) / (math.sqrt(((1-((eta)**2))**2)+((2*eta*D)**2))))

A_s = 1.0
eta = np.linspace(0,3,1000)
D = [0.1, 1.0, 3.0]

A1 = np.copy(eta)
A2 = np.copy(eta)
A3 = np.copy(eta)

for i in range(len(eta)):
    A1[i] = A(eta[i], A_s, D[0])
    A2[i] = A(eta[i], A_s, D[1])
    A3[i] = A(eta[i], A_s, D[2])

fig, axes = plt.subplots()
axes.plot(eta, A1, label=f"D = {D[0]}")
axes.plot(eta, A2, label=f"D = {D[1]}")
axes.plot(eta, A3, label=f"D = {D[2]}")
axes.legend(loc=1)
axes.set_xlabel(r'$\eta$', fontsize=14)
axes.set_ylabel("Amplitude")
axes.set_title("Resonance curves with different parameters D");
```
