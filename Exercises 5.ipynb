```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-(np.pi), np.pi, num=1000)
f = 0.5 + ((5*10**(-2))*x * (x + 2.5) * (x - 3))
g = np.sin(49*x)

fig, axes = plt.subplots()

axes.plot(x, f, '--k', label="f(x)")
axes.plot(x, g, '-g', label="g(x)", alpha=0.42)

axes.legend(loc=2)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Task 1')

fig.tight_layout()
```

### Task 2:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def move_avg(data, r):
    try:
         if r <= 0 or type(r) != int:
             raise Exception
    except Exception:
        print("Make sure the radius is an integer greater than 0. Please try again.")
    else:
        avg_data = []
        for k in range(len(data)):
            if k-r < 0 or k+r >= len(data):
                avg_data.append(None)
            else:
                avg = (sum(data[k-r:k+r]))/(2*r)
                avg_data.append(avg)
        return avg_data

sun = np.loadtxt("sunspots.txt")
mon = sun[:,0]
spots = sun[:,1]

## Part 1

fig1, axes1 = plt.subplots()
axes1.plot(mon, spots)
axes1.set_xlabel('Month after Jan 1749')
axes1.set_ylabel('Sunspots')
axes1.set_title('Sunspots /month from Jan 1749');
fig1.tight_layout(pad=5)

## Part 2

mon_short = sun[:999,0]
spots_short = sun[:999,1]

fig2 = plt.figure()
axes2_1 = fig2.add_axes([0, 0, 1, 1])
axes2_2 = fig2.add_axes([0.6, 0.7, 0.3, 0.2])
axes2_1.plot(mon_short, spots_short)
axes2_1.set_xlabel('Month after Jan 1749')
axes2_1.set_ylabel('Sunspots')
axes2_1.set_title('Sunspots /month in first 1000 months after Jan 1749')
axes2_2.plot(mon, spots, '-r')
axes2_2.set_xlabel('Months after Jan 1749')
axes2_2.set_ylabel('Sunspots')
axes2_2.set_title('Sunspots /month from Jan 1749');

## Part 3

radius5 = move_avg(spots, 5)
radius50 = move_avg(spots, 50)
radius100 = move_avg(spots, 100)

fig3, axes3 = plt.subplots(4, 1)
axes3[0].plot(mon, spots)
axes3[0].set_xlabel('Month after Jan 1749')
axes3[0].set_ylabel('Sunspots')
axes3[0].set_title('Sunspots /month from Jan 1749');
axes3[1].plot(mon, radius5, '-r')
axes3[1].set_xlabel('Month after Jan 1749')
axes3[1].set_ylabel('Moving average')
axes3[1].set_title('Moving average of sunspots /month with radius 5');
axes3[2].plot(mon, radius50, '-y')
axes3[2].set_xlabel('Month after Jan 1749')
axes3[2].set_ylabel('Moving average')
axes3[2].set_title('Moving average of sunspots /month with radius 50');
axes3[3].plot(mon, radius100, '-g')
axes3[3].set_xlabel('Month after Jan 1749')
axes3[3].set_ylabel('Moving average')
axes3[3].set_title('Moving average of sunspots /month with radius 100');
fig3.tight_layout()
```
